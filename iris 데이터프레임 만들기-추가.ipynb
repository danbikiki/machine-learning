```
import pandas as pd
import pandas as pd
from sklearn.datasets import load_iris   
# sklearn.datasetsincludes common example datasets
# A function to load in the iris dataset
iris=load_iris()
iris_data=iris.data
iris_data
```

```
iris_label=iris.target
iris_label
```

```
import pandas as pd
df_iris=pd.DataFrame(iris_data)
df_iris.columns=['sepal length','sepal width','petal length','petal width']
df_iris
```

```
s_target=pd.Series(iris_label)
s_target
```

```
df_iris['species']=s_target
df_iris
```
