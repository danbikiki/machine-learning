```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```
tips=sns.load_dataset('tips')
tips
```

```
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='sex')
```

```
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker')
```

```
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```
sns.relplot(x='total_bill',y='tip',hue='size',data=tips,palette='ch:r=-.5,.75')
```

```
sns.relplot(x='total_bill',y='tip',data=tips,size='size')
```

```
df=pd.DataFrame(dict(time=np.arange(500),
                    value=np.random.randn(500).cumsum()))
```

```
g=sns.relplot(x='time',y='value',kind='line',data=df)
g.fig
```
